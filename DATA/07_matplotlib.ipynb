## 07 matplotlib
- 파이썬에서 가장 널리 사용되는 데이터 시각화 라이브러리

```python
import matplotlib
import matplotlib.pyplot as plt # matplotlib의 서브 패키지
import numpy as np

matplotlib.__version__
```

```python
# 한글 깨짐 방지

# Windows에서 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # '맑은 고딕'이 설치되어 있을 경우
plt.rcParams['axes.unicode_minus'] = False     # 마이너스(-) 부호 깨짐 방지
```

```python
# macOS에서 한글 폰트 설정
plt.rcParams['font.family'] = 'AppleGothic'   # macOS 기본 한글 폰트
plt.rcParams['axes.unicode_minus'] = False
```

### 그래프 그리기

```python
# 그래프 그리기
x = [1,2,3,4,5]
y = [10,14,19,23,25]

# 선 그래프
plt.plot(x,y)

# 이미지를 파일로 저장
# plt.savefig("my_graph.png")

# 콘솔에 이미지 그리기
plt.show()
```

### 그래프의 구성 요소 (1)
- figure : 여러 그래프가 그려지는 최상위 컨테이너
- axes : 실제 그래프가 그려지는 공간
- axis : 그래프의 축

```python
# subplots()

# fig, axs = plt.subplots(행, 열)
fig, axs = plt.subplots() # 행과 열을 입력하지 않으면 하나의 figure와 하나의 axes를 만듦
```

```python
# 1행 2열의 axes 생성
fig, axs = plt.subplots(1,2)
axs[0].plot([1,2,3], [1,2,3])
axs[1].plot([1,2,3], [1,4,9])

plt.show()
```

```python
# 2행 2열의 axes 생성
fig, axs = plt.subplots(2,2)

axs[0,0].plot([1,2,3], [1,2,3])
axs[0,1].plot([1,2,3], [1,4,9])
axs[1,0].plot([1,2,3], [3,2,1])
axs[1,1].plot([1,2,3], [9,4,1])
# plt.tight_layout() # 그래프 간격 자동 조정

plt.show()
```

### 그래프의 구성 요소 (2)
- 제목, 레이블, 범례, 그리드

```python
# title - 기본 방식
# 아래처럼 축약형으로 가능
# plt.plot([1,2,3], [1,4,9], "bo--", label="제곱 그래프")
plt.plot([1,2,3], [1,4,9], label="제곱 그래프", linestyle=":", color="g", marker="o")
plt.title("제곱 함수 그래프", color="b", fontsize=20)

# x label, y label
plt.xlabel("x 값")
plt.ylabel("y 값(x의 제곱)")

# xticks, yticks
plt.xticks([1,2,3], ["A", "B", "C"])
plt.yticks([1,5,10,15])

# 범례 (legend) 
# 생성시 label= 설정 필수
plt.legend()

# 격자 (grid)
plt.grid(linestyle="--", alpha=0.5)

plt.show()
```

```python
# title - subplots
fig, axs = plt.subplots()

axs.plot([1,2,3],[1,4,9], label="제곱 그래프")
axs.set_title("제곱 함수 그래프")

# x label, y label
axs.set_xlabel("x 값")
axs.set_ylabel("y 값(x의 제곱)")

# 범례 (legend) 
# 생성시 label= 설정 필수
axs.legend()

# 격자 (grid)
axs.grid()

plt.show()
```

```python
fig, axs = plt.subplots(2,2)

fig.suptitle("여러 함수 그래프") # 전체 이름 설정

axs[0,0].plot([1,2,3], [1,2,3])
axs[0,0].set_title("1번 그래프")

axs[0,1].plot([1,2,3], [1,4,9])
axs[0,1].set_title("2번 그래프")

axs[1,0].plot([1,2,3], [3,2,1])
axs[1,0].set_title("3번 그래프")

axs[1,1].plot([1,2,3], [9,4,1])
axs[1,1].set_title("4번 그래프")

plt.tight_layout() # 그래프 간격 자동 조정

plt.show()
```

```python
# 하나의 axes에 여러개의 그래프 그리기

x = np.arange(10)
y1 = 2 * x + 1
y2 = 3 * x - 5

plt.plot(x, y1, marker="o", label="y = 2x + 1", color="c")
plt.plot(x, y2, marker="*", label="y = 3x - 5", color="m")

plt.title("두개의 선 그래프 비교")

plt.xlabel("x값")
plt.ylabel("y값")
plt.legend()
plt.grid(linestyle=":", alpha=0.5)

plt.show()
```

```python
# 실습 1. 선 그래프 그리기
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.plot(months, sales_2019, label="2019", linestyle=":", color="b", marker="o")
plt.plot(months, sales_2020, label="2020", linestyle=":", color="r", marker="D")

plt.title("Monthly Sales Comparison (2019-2020)", fontsize=15, color="g")

plt.grid(linestyle="--", alpha=0.3)

plt.legend(loc="lower right", title="판매 연도")

plt.xlabel("Month", labelpad=15)
plt.ylabel("Sales", labelpad=15)

plt.tight_layout()

plt.show()
```

## 다양한 2D 그래프 그리기

```python
# 수직 막대 그래프
labels = ["A", "B", "C", "D"]
values = [10,24,15,32]

plt.bar(labels, values, width=0.5)

plt.show()
```

```python
# 수평 막대 그래프
labels = ["A", "B", "C", "D"]
values = [10,24,15,32]

plt.barh(labels, values, height=0.5)

plt.show()
```

```python
# 여러 막대 그래프
labels = ["1분기", "2분기", "3분기", "4분기"]
men_means = [20,34,30,35]
women_means = [25,32,34,40]

x = np.arange(len(labels))

fig, ax = plt.subplots()
width = 0.4

ax.bar(x - width/2, men_means, width, label="남성")
ax.bar(x + width/2, women_means, width, label="여성")

ax.legend()

ax.set_xticks(x, labels)

plt.show()
```

```python
# 실습 2. 막대 그래프 그리기
data = [20, 35, 15, 27, 45]
categories = ["Category 1", "Category 2", "Category 3", "Category 4", "Category 5"]

plt.bar(categories, data, width=0.5, color="k", label="Value")

plt.title("Bar Chart", fontsize=10)

plt.grid(linestyle=":", alpha=0.7, axis="y")

plt.legend(loc="upper left")

plt.xlabel("Categories", labelpad=15)
plt.ylabel("Values", labelpad=15)

plt.xticks(rotation=45)

plt.ylim(0,50)

plt.show()
```

### 히스토그램 (Histogram) 그리기

```python
# 히스토그램

# 랜덤 데이터 생성
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor="k")

plt.show()
```

```python
# 실습 3. 히스토그램 그리기
# cnt, bins = plt.hist(data, bins=(np.arange(1,8) - 0.5))
np.random.seed(42)
data = np.random.randint(1,7,(100))

plt.hist(data, bins=(np.arange(1,8) - 0.5), edgecolor="k", color="g", rwidth=0.5)

plt.title("주사위 눈 빈도 히스토그램")

plt.grid(linestyle=":", alpha=0.7, axis="y")

plt.xlabel("주사위 눈")
plt.ylabel("빈도")

plt.ylim(0,30)

plt.show()
```

### 산점도 그리기

```python
# 산점도 그리기

# x = [5,7,8,10,12]
# y = [20,15,25,35,30]

x = np.linspace(2, 30, 20)
y = np.linspace(10, 40, 20)
y_noisy = y + np.random.uniform(-7, 7, 20)

plt.scatter(x, y_noisy)
plt.grid(linestyle=":")

plt.show()
```

```python
# 다차원 데이터 표현
np.random.seed(42)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)

colors = np.random.rand(n)
sizes = np.random.rand(n) * 1000

plt.scatter(x, y, s=sizes, c=colors, alpha=0.7, cmap="viridis")
plt.colorbar()

plt.show()
```

```python
# 실습 4. 산점도 그리기
np.random.seed(42)

x = np.linspace(0, 10, 30)
y = 3 * x + 2 + np.random.normal(0, 3, 30)

colors = np.random.rand(30)
sizes = np.random.rand(30) * 100

plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, marker=",")

plt.title("노이즈가 추가된 산점도")

plt.grid(linestyle=":", alpha=0.7)

plt.xlabel("x 값")
plt.ylabel("y 값", rotation=0)

plt.show()
```

### 파이차트

```python
# 파이차트
labels = ["A", "B", "C", "D"]
sizes = [40,20,25,15]
explode = [0.1, 0, 0, 0]

plt.pie(sizes, labels=labels, autopct="%1.0f%%", shadow=True, explode=explode, startangle=60)

plt.show()
```

```python
# 실습 5. 파이 차트 그리기
labels = ["Apple", "Banana", "Mango", "Blueberry"]
sizes = [15,30,45,10] 
explode = [0,0.1,0,0]

plt.pie(sizes, labels=labels, autopct="%1.0f%%", shadow=True, explode=explode, startangle=80, colors=["r", "y", "m", "b"])

plt.show()
```
